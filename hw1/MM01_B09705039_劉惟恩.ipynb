```python
""""
question 1_(a)

Purpose : Using numpy to build matrix A and vector b.
Result : matrix A :[[3 2 0]      vector b:[[ 2]
                    [1 1 0]                [ 4]
                    [0 5 1]]               [-1]]
"""""
matrix_A = np.array([[3,2,0],
                    [1,1,0],
                    [0,5,1]])
print(matrix_A)

vector_b = np.array([[2],
                    [4],
                    [-1]])
print(vector_b)
```

```python
""""
question 1_(b)

Purpose : Using numpy.linalg to solve the question Ax = b.
Result : x = [[ -6.]
              [ 10.]
              [-51.]]
"""""
from numpy.linalg import solve

x = solve(matrix_A, vector_b)
print(x)
```

```python
""""
question 1_(c)

Purpose : Using numpy.linalg to find the inverse matrix of matrix A.
Result : A inverse = [[  1.  -2.   0.]
                      [ -1.   3.   0.]
                      [  5. -15.   1.]]
"""""
A_inverse = np.linalg.inv(matrix_A)
print(A_inverse)
#print(np.dot(matrix_A, A_inverse)) to check if it is the inverse of matrix A
```

```python
""""
question 1_(d)

Purpose : Using numpy to calcutate matrix A inverse times vector b.
Result : result = [[ -6.]   is equal to the answer of x in Ax = b, which is resonable since to solve Ax = b we multiply both side with A inverse.
                   [ 10.]
                   [-51.]]
"""""
y = np.dot(A_inverse, vector_b)
print(y)
```

```python
#imports required
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import cv2
from numpy import asarray
#import numpy as np  (is defined at the first cell of this program)
import sys
%matplotlib inline
```

```python
""""
question 2_(a)

Purpose : Using "Matrix Addition" to decorate the Dr. Lee’s picture.
Result : After doing the process below it will output the decorated picture as Addition.jpg.
"""""
img1 = cv2.imread('pic1.jpg') # read an image file in the program
img2 = cv2.imread('pic2.jpg')
img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0])) # resize the height and width of pic2

img_add = cv2.add(img1,img2_resized) # add two image matrices together

cv2.imwrite('Addition.jpg',img_add) # output the image matrices into a file
addition = Image.open('Addition.jpg') # open the output file
plt.imshow(addition) # display the image on the screen
#(since using cv2.imshow('Subtraction',img),cv2.waitKey(0),cv2.destroyAllWindows() somehow crashes the computer so I changed the way to output the image)
```

```python
""""
question 2_(b)

Purpose : Using "Matrix Subtraction" to decorate the Dr. Lee’s picture.
Result : After doing the process below it will output the decorated picture as Subtraction.jpg.
"""""
#the file reading process is same as the previous cell so it's no need to be written again

img_sub = cv2.subtract(img1, img2_resized) # subtract two image matrices

cv2.imwrite('Subtraction.jpg',img_sub) # output the image matrices into a file
subtraction = Image.open('Subtraction.jpg') # open the output file
plt.imshow(subtraction) # display the image on the screen
```

```python
""""
question 2_(c)

Purpose : Using opencv to rotate Dr. Lee’s picture (w.r.t) center to 45 degree without scaling.
Result : After doing the process below it will output the picture as Rotate.jpg.
"""""
#the file reading process is same as the previous cell so it's no need to be written again

(rows, cols) = img1.shape[:2] # Shape of image in terms of pixels.

M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1) # getRotationMatrix2D creates a matrix needed for transformation.
img_rot = cv2.warpAffine(img1, M, (cols, rows)) # after multiplying the rotating matrix

cv2.imwrite('Rotate.jpg',img_rot) # output the image matrices into a file
rotate = Image.open('Rotate.jpg') # open the output file
plt.imshow(rotate) # display the image on the screen
```

```python
""""
question 2_(d)

Purpose : Using Canny Edge Detection (in Open CV) for Dr. Lee’s picture.
Result : After doing the process below it will output the picture as EdgeDetect.jpg.
"""""
#the file reading process is same as the previous cell so it's no need to be written again

img_edge = cv2.Canny(img1, 100, 200) # use Canny Edge Detection to fransform the picture

cv2.imwrite('EdgeDetect.jpg',img_edge) # output the image matrices into a file
edgeDetect = Image.open('EdgeDetect.jpg') # open the output file
plt.imshow(edgeDetect) # display the image on the screen
```
